```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('train.txt' , sep=';' , header= None , names =['text','emotion'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
unique_emotions = df['emotion'].unique()
#create a dictionary to save emotions 
emotion_numbers ={} 
#give numeric values to the emotions 
i = 0
for emo in unique_emotions:
    emotion_numbers[emo] = i 
    i+=1 

df['emotion'] = df['emotion'].map(emotion_numbers)
```

```python
df
```

```python
#converting all texts in lower case 

df['text'] = df['text'].apply(lambda x:x.lower())
```

```python
#remove puncuations :-
import string 
def remove_punc(text):
    return text.translate(str.maketrans('','' ,string.punctuation))
    
```

```python
df['text'] = df['text'].apply(remove_punc)
```

```python
#now remove numbers 
def remove_numbers(text):
    new =""
    for i in text:
        if not i.isdigit():
            new = new + i
    return new 
df['text'] = df['text'].apply(remove_numbers)    
```

```python
df
```

```python
#remove urls and links 
import re

def remove_urls(text):
    if isinstance(text, str):
        return re.sub(r'http\S+|www\S+', '', text)
    return text
df['text'] = df['text'].apply(remove_urls)
```

```python
#remove emojies and special characters :-
def remove_emojis(text):
    new = ""
    for i in text:
        if i.isascii():
            new += i 
    return new 

df['text']=df['text'].apply(remove_emojis)
```

```python
#remove stopwords :-
import nltk 
```

```python
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
```

```python
nltk.download("punkt")   #used for tokenization
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
stop_words
```

```python
len(stop_words)
```

```python
#tokenization 
def remove(text):
    if isinstance(text, str):
        words = word_tokenize(text)
        cleaned = []
        for i in words:
            if i not in stop_words:
                cleaned.append(i)
        return ' '.join(cleaned)
    return text
```

```python
nltk.download('punkt_tab')
```

```python
df['text'] = df['text'].apply(remove)
```

```python
df.loc[1]['text']
```

```python
#now lets split the text means make text and emotion seprate in ratio of 80:20 

from sklearn.model_selection import train_test_split

x=df['text']
y=df['emotion']

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.20, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
#CountVectorizer is for bag of words which contain only 0 and 1 
# Tfidf is for term frequencies IDFs ko calculate krta hai aur uske basis pe apaka points m,e deta hai 
```

```python
#1st model using bag of words 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer , CountVectorizer

from sklearn.naive_bayes import MultinomialNB 
from sklearn.metrics import accuracy_score

bow_vactorizer = CountVectorizer()
x_train_bow = bow_vactorizer.fit_transform(x_train)
x_test_bow = bow_vactorizer.transform(x_test)
#using navbias model 
nb_model = MultinomialNB()
nb_model.fit(x_train_bow,y_train)

pred_bow = nb_model.predict(x_test_bow)
print("Accuracy:", accuracy_score(y_test,pred_bow))
```

```python
pred_bow 
```

```python
#now do by using TfidfVectorizer
```

```python
Tfidf_Vectorizer  =  TfidfVectorizer()

x_train_tfidf = Tfidf_Vectorizer.fit_transform(x_train)
x_test_tfidf = Tfidf_Vectorizer.transform(x_test)


nb2_model = MultinomialNB ()
nb2_model.fit(x_train_tfidf,y_train)
```

```python
#lets predict 
y_pred = nb2_model.predict(x_test_tfidf)
print("Accuracy:", accuracy_score(y_test,y_pred))
```

```python
#lets use logistic regression using tf-idf and make a model 

from sklearn.linear_model import LogisticRegression
```

```python
logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(x_train_tfidf,y_train)
```

```python
#now test 
log_pred = logistic_model.predict(x_test_tfidf)
```

```python
print("Accuracy:" , accuracy_score(y_test,log_pred))
```
